# Protein pLLPS Score Lookup

This notebook allows you to quickly look up pLLPS scores for a list of protein names.

```python
import pandas as pd
from llps_functions import is_membrane_protein
```

```python
# Load the dataset
df = pd.read_csv('results/full_dataset.csv')
print(f"Dataset loaded: {len(df)} proteins")
```

## Look up pLLPS scores

**Simply paste your list of protein names below** (one per line) and run the cell to get their pLLPS scores.

```python
# Paste your protein names below (one per line). Mix of Entry names or Protein names is fine.
names_text = """
IL4_HUMAN
SOCS5_HUMAN
JAK2_HUMAN
I13R2_HUMAN
IL13_HUMAN
JAK1_HUMAN
STAT6_HUMAN
JAK3_HUMAN
IL2RG_HUMAN
IL4RA_HUMAN
I13R1_HUMAN
"""

# Turn the pasted block into a clean list
protein_names = [line.strip() for line in names_text.splitlines() if line.strip()]

# Look up scores in Protein names OR Entry name
results = []
for name in protein_names:
    match = df[
        df['Protein names'].str.contains(name, case=False, na=False)
        | df['Entry name'].str.contains(name, case=False, na=False)
    ]
    if not match.empty:
        first = match.iloc[0]
        score = first['p(LLPS)']
        full_name = first['Protein names']
        entry_name = first['Entry name']
        
        # Check if membrane protein
        is_membrane = is_membrane_protein(
            function_str=first.get('Function [CC]', ''),
            protein_name_str=full_name,
            location_str=first.get('Subcellular location [CC]', '')
        )
        
        results.append({
            'Query': name,
            'pLLPS_Score': score,
            'Is_Membrane': is_membrane,
            'Entry_Name': entry_name,
            'Full_Name': full_name,
        })
    else:
        results.append({
            'Query': name,
            'pLLPS_Score': 'Not found',
            'Is_Membrane': 'N/A',
            'Entry_Name': 'N/A',
            'Full_Name': 'N/A',
        })

# Display as DataFrame
results_df = pd.DataFrame(results)
results_df
```

## Export results (optional)

Uncomment the line below to save results to a CSV file.

```python
results_df.to_csv('results/protein_pllps_lookup.csv', index=False)
print("Results saved to results/protein_pllps_lookup.csv")
```
